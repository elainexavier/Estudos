```python
import pandas as pd
```

### 1.1 Importando um txt

```python
url = 'https://raw.githubusercontent.com/jovianosilveira/materiais_apoio_cursos/main/Palestras/202104%20-%20Guia%20Essencial%20Python/dados/mussum2.txt'
df = pd.read_csv(url, sep = '|', engine= 'c')
df
```

> Ai você percebeu que ele "não mostra tudo", e para resolver isso, temos que setar as opções da biblioteca

```python
pd.set_option('max_colwidth', 400)
df
```

> suponhamos que você queira somente as Frases com ID >= 5

```python
df.query('FraseID >= 5 ')
```

> Agora vamos tentar filtrar a FraseData >= '01/01/1990'

```python
df.query('FraseData >= "01/01/1990" ')
```

```python
df.dtypes
```

```python
df['FraseData']
```

```python
df.FraseData
```

```python
# mostrando qual seria o resultado da serie se executada a função
pd.to_datetime(df['FraseData'])
```

```python
# atribuindo à serie e função anterior, ou seja, alterando a coluna original
df['FraseData'] = pd.to_datetime(df['FraseData'])
```

```python
df
# parece que nada mudou, mas...
```

```python
df.query('FraseData >= "01/01/1990" ')
```

```python
# E se eu quisesse que fosse uma variável (eu prefiro)
DataCorte = "01/01/1990"
df.query('FraseData >= @DataCorte ')
```

### 1.2 Importando um excel

```python
url = 'https://github.com/jovianosilveira/materiais_apoio_cursos/blob/main/Palestras/202104%20-%20Guia%20Essencial%20Python/dados/mussum.xlsx?raw=true'
```

```python
# Faça, e perceba a diferença...
df2 = pd.read_excel(url)
df2
```

```python
df2.dtypes
```

```python
# Então...eu posso
DataCorte = "01/01/1990"
df2.query('FraseData >= @DataCorte ')
```

> Suponhamos que você queira os 5 primeiros caracteres da frase.
>
> Em uma variável você faria:

```python
frase = 'Mussum Ipsum, cacilds vidis litro abertis. Mais vale um bebadis conhecidiss, que um alcoolatra anonimis. Diuretics paradis num copo é motivis de denguis. Atirei o pau no gatis, per gatis num morreus. Casamentiss faiz malandris se pirulitá.'
frase[:5]
```

> Mas no dataframe, você tem que cuidar...

```python
df2['Teste'] = df2['FraseDescricao'][:5]
df2
```

> Para você interagir com cada linha do dataframe, você precisa usar um Lambda

```python
df2['Teste'] = df2['FraseDescricao'].apply(lambda x: x[:5])
df2
```

```python
# Se quiser usar mais de uma coluna
df2['Teste2'] = df2[['FraseDescricao','FraseID']].apply(lambda x: x['FraseDescricao'][:5] + str(x['FraseID']) , axis=1)
df2
```

### 1.3 Usando função personalizada na coluna

> Uma função personalizada, é um conjunto de ações que sempre vai executar da mesma maneira sempre que executadas.
>
> Uma biblioteca é um livro de funções personalizadas.
>
> Vamos transformar o exemplo anterior em uma função personalizada

```python
def concatena(a,b):
    c = str(a) + str(b)
    return c
```

```python
# Se quiser usar mais de uma coluna
df2['Teste3'] = df2[['FraseDescricao','FraseID']].apply(lambda x: concatena(  x['FraseDescricao'][:5], str(x['FraseID'])  ) , axis=1)
df2
```
